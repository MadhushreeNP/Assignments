```
#creating a dictionary
country_info = { "Name": "India","capital": "Delhi"}
print(country_info)
```

```
country_info = { "Name": "India","capital": "Delhi"}
print(country_info["Name"])
```

```
country_info = { "Name": "India","capital": "Delhi"}
print(country_info["capital"])
```

```
#creating a dictionary
country_info = { "Name": "Japan","capital": "tokyo"}
print(country_info)
print(country_info["Name"])
print(country_info["capital"])
```

```
country_info[ 'population'] = '139.67M'
print(country_info)
```
